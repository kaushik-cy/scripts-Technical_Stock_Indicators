```python
import stonkslab as sl
import stonksplot as sp
```

```python
stock = 'SBIN.NS'
start_date = '2015-01-01'
end_date = '2021-02-28'
```

```python
data = sl.download_data(stock, start_date, end_date)
```

```python
data_i = sl.all_indicators(data)
data_i.tail()
```

```python
sp.DIMS = (15,7.5)
```

# Change

```python
sp.plot_change(data_i, 260)
```
